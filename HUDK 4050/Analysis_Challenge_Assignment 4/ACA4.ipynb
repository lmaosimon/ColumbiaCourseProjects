###Implementation###
Creating a network using NetworkX.

```python
#Using networkx library
import networkx as nx

#Build the graph
G_symmetric = nx.Graph()
G_symmetric.add_edge('Shelly', 'Simon')
G_symmetric.add_edge('Shelly', 'Jianan Dingqian')
G_symmetric.add_edge('Shelly', 'Luke')
G_symmetric.add_edge('Shelly', 'Andy')
G_symmetric.add_edge('Luke', 'Jianan Dingqian')
G_symmetric.add_edge('Luke', 'Simon')
G_symmetric.add_edge('Luke', 'Andy')
G_symmetric.add_edge('Jianan Dingqian', 'Simon')
```

```python
#visualize the network
nx.draw_networkx(G_symmetric)
```

How many classes we are having together?
>We are using the weighted Network to show it.

```python
G_weighted = nx.Graph()
G_weighted.add_edge('Shelly', 'Simon', weight = 3)
G_weighted.add_edge('Shelly', 'Jianan Dingqian', weight = 2)
G_weighted.add_edge('Shelly', 'Luke', weight = 2)
G_weighted.add_edge('Shelly', 'Andy', weight = 1)
G_weighted.add_edge('Luke', 'Jianan Dingqian', weight = 3)
G_weighted.add_edge('Luke', 'Simon', weight = 3)
G_weighted.add_edge('Luke', 'Andy', weight = 2)
G_weighted.add_edge('Jianan Dingqian', 'Simon', weight = 3)
nx.draw_networkx(G_weighted)
```

The social network has tendency to form clusters.<br/>
We are trying to get average clustering corfficient for the classmake network

```python
nx.average_clustering(G_symmetric)
```

We noted that the distance of a node from every other node in the network using BFS algorithm, starting from that node.

```python
T = nx.bfs_tree(G_symmetric, 'Jianan Dingqian')
nx.draw_networkx(T)
```

Using eccentricity to find the largest distance between A and all other nodes.

```python
nx.eccentricity(G_symmetric)
```

We find that the center in this graph is Shelly and Luke who have the most influence in this network.

```python
nx.center(G_symmetric)
```
